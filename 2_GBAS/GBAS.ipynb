# Sistemas de Aumentacion GBAS: Ground Based Augmentation systems

### ¿Que son sistemas de aumentacion?

Ayudan a monitorear los sistemas de posicionamiento global (GPS, GLONASS) y proveer correcciones mejorando la calidad del servicio.

- Mejoran la **integridad** con monitoreo en tiempo real.
- Mejoran la **precisión** por medio de correcciones diferenciales.
- Mejoran la **disponibilidad** y **continuidad.**

Existen 2 tipos de sistemas de aumentación: 

- SBAS (Con satélites geoestacionarios)
- GBAS (Con ayuda de estaciones terrestres)

###  Por qué se necesitan?

**En aviación**  desempeño de los GNSS no cumple con todos los requisitos para las fases de vuelo. Los usuarios de tierra y mar también necesitan de servicios de aumentación.


# GBAS

En un GBAS  se tiene una estacion terrestre de referencia para mejorar el desempeño de los servicios de navegación en area local ya que el SPS (Standard positioning service) por si solo no cumple con los requerimientos de aviacion . Se usa para dar soporte en todas las fases de vuelo: despegue, aterrizaje. 
 
### Segmentos de un GBAS

- Terrestre (estacion que provee correcciones al usuario)
- Espacial (constelacion de satélites (GPS o GLONASS)
- Segmento aéreo (usuario)

Las correcciones que hace se emiten a traves de una canal de datos de alta frecuencia VHF --> VDB
La estacion puede proporcionar posicionamiento a usuarios que se encuentren hasta 100 kilómetros. La distancia entre estacion y usuario (baseline) muy grande puede degradar las mediciones.

- VHF Very high frequency
- VDB: VHF Data Broadcast 
- VOR : VHF omnidirectional ranging

navigation system **[5]. As such, the basic principle is that
pseudorange observations made by ground-based receivers
are used to develop differential corrections for each
satellite. These corrections are provided to the airborne
user’s receiver via a data link. The airborne receiver then
applies these corrections in order to produce a set of
corrected pseudoranges that are then the basis of a position
solution. The underlying assumption is that, for relatively
short baseline separations between the ground-based
receivers and the airborne-based receivers, the most
significant error sources will be common to both observers
and will therefore be eliminated by the differential
processing [6]
